```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Load data
train = pd.read_csv('/kaggle/input/titanic/train.csv')
test = pd.read_csv('/kaggle/input/titanic/test.csv')

# Drop columns with many missing values or not useful
train = train.drop(['Cabin', 'Ticket'], axis=1, errors='ignore')
test = test.drop(['Cabin', 'Ticket'], axis=1, errors='ignore')

# Fill missing values for Age and Fare
train['Age'] = train['Age'].fillna(train['Age'].median())
test['Age'] = test['Age'].fillna(test['Age'].median())
test['Fare'] = test['Fare'].fillna(test['Fare'].median())

# Fix for Embarked column fill
if train['Embarked'].mode().empty:
    fill_value = 'S'  # fallback if mode() is empty
else:
    fill_value = train['Embarked'].mode()[0]
train['Embarked'] = train['Embarked'].fillna(fill_value)

# Convert categorical columns to numeric
train['Sex'] = train['Sex'].map({'male': 0, 'female': 1})
test['Sex'] = test['Sex'].map({'male': 0, 'female': 1})

train['Embarked'] = train['Embarked'].map({'S':0, 'C':1, 'Q':2})
test['Embarked'] = test['Embarked'].map({'S':0, 'C':1, 'Q':2})

# Select features and target
features = ['Pclass', 'Sex', 'Age', 'Fare', 'Embarked']
X = train[features]
y = train['Survived']
X_test = test[features]

# Train Decision Tree model
model = DecisionTreeClassifier()
model.fit(X, y)

# Predict on test set
predictions = model.predict(X_test)

# Prepare submission file
submission = pd.DataFrame({
    'PassengerId': test['PassengerId'],
    'Survived': predictions
})

submission.to_csv('submission.csv', index=False)

print("Model trained and submission file created successfully!")
```

```python
# Step 1: Import necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from IPython.display import FileLink

# Step 2: Load the data
train_df = pd.read_csv("/kaggle/input/titanic/train.csv")
test_df = pd.read_csv("/kaggle/input/titanic/test.csv")

# Step 3: Preprocess data
train_df["Age"].fillna(train_df["Age"].median(), inplace=True)
test_df["Age"].fillna(test_df["Age"].median(), inplace=True)
train_df["Fare"].fillna(train_df["Fare"].median(), inplace=True)
test_df["Fare"].fillna(test_df["Fare"].median(), inplace=True)
train_df["Embarked"].fillna("S", inplace=True)
test_df["Embarked"].fillna("S", inplace=True)

# Convert categorical to numeric
for df in [train_df, test_df]:
    df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
    df["Embarked"] = df["Embarked"].map({"S": 0, "C": 1, "Q": 2})

# Step 4: Select features
features = ["Pclass", "Sex", "Age", "Fare"]
X = train_df[features]
y = train_df["Survived"]
X_test = test_df[features]

# Step 5: Train the model
model = DecisionTreeClassifier()
model.fit(X, y)
predictions = model.predict(X_test)

# Step 6: Create submission file
submission = pd.DataFrame({
    "PassengerId": test_df["PassengerId"],
    "Survived": predictions
})
submission.to_csv("/kaggle/working/submission.csv", index=False)

# Step 7: Provide download link
FileLink(r'/kaggle/working/submission.csv')

```
